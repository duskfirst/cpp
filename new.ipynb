```python
time = linspace(0.0, 10.0, 100)

height = exp(-time/3)*sin(time * 3)
figure()
plot(time, height, "m-^")

plot(time, 0.3 * sin(time*3), "g-")
xlabel('Time(s)')
ylabel('Height')
xlim(0, 8)
legend(['damped', 'constant amplitude'], loc ='upper right')
show()
```

```python
def plottrig(f):
    xvalue = linspace(-pi, pi, 100)
    plot(xvalue, f(xvalue), "r-")
    xlim(-pi, pi)
    ylim(-20, 20)
trigf= (sin, cos, tan)
for i in trigf:
    print(i(pi/6))
    plottrig(i)
show()
```
